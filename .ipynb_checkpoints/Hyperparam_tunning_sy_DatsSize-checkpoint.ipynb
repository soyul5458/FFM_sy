```python
# 범례 이름 출력하기 위해 변경
def ModelRename(x):
    if x == "BCResMax_LA_":
        return "BC-ResMax" 
    elif x == "DDWSseq_LA_":
        return "DDWS"
    elif x == "LCNN_LA_":
        return "LCNN"
    elif x == "ResMax_LA_":
        return "ResMax"
    else:
        return x
```

```python

def Percentage(x):
    return x*100
```

```python
import numpy as np

import os
import pandas as pd
import pickle
#from models.BCresnet2 import get_BCResNet, get_inputshape
from sklearn.metrics import roc_curve
from utils.DataLoader import data_loader
from utils.Generator0 import DataGenerator, feature_extract_cqt, evalEER,  evalScore, evalEER_f, evalEER_f2, gen_fname


reslist = ['res_LA_DataSize/'+fnm for fnm in os.listdir('res_LA_DataSize/')]

data = []
allkeys = []
for i in range(len(reslist)) :
    with open(reslist[i], 'rb') as f:
        data1 = pickle.load(f)
    data.append(data1)
    allkeys += data1.keys()

allkeys = list(set(allkeys))

newlist = {}
for i in range(len(data)) :
    if i == 0 :
        for k1 in allkeys :
            try :
                newlist[k1] = [ data[i][k1] ]
            except :
                newlist[k1] = [-999]
    else :
        for k1 in allkeys :
            try :
                newlist[k1].append(data[i][k1])
            except :
                newlist[k1].append(-999)

df1 = pd.DataFrame(newlist)
df1.eer_val = pd.to_numeric(df1.eer_val)
df1.eer_eval = pd.to_numeric(df1.eer_eval)
```

## None

```python
df1[df1.beta_param== False][df1.highpass == False][df1.lowpass == False][df1.ranfilter2 == False][['model','DataSize','beta_param','lowpass','highpass','ranfilter2','eer_val','eer_eval']].sort_values( ['model','DataSize','eer_eval'] )
```

```python
df1_1 = df1[df1.beta_param== False][df1.highpass == False][df1.lowpass == False][df1.ranfilter2 == False][['model','DataSize','beta_param','lowpass','highpass','ranfilter2','eer_val','eer_eval']].groupby(['model','DataSize'])['eer_eval'].agg(**{
    'min_eer_eval':'min',
    'mean_eer_eval':'mean',
    'std_eer_eval' :'std'
}).reset_index()
df1_1
```

```python
df1_2 = df1_1.copy()
df1_2['model'] = df1_2['model'].apply(ModelRename)
df1_2['DataSize'] = df1_2['DataSize'].apply(Percentage)
df1_2['mean_eer_eval'] = df1_2['mean_eer_eval'].apply(Percentage)
```

```python
df1_2.to_csv("DataSize_None.csv", index = False)
```

## Highpass

```python
df1[df1.highpass != False][df1.lowpass == False][df1.ranfilter2 == False][['model','DataSize','lowpass','highpass','ranfilter2','eer_val','eer_eval']].sort_values( ['model','DataSize','eer_eval'] )
```

```python
df2 = df1[df1.highpass != False][df1.lowpass == False][df1.ranfilter2 == False][['model','DataSize','beta_param','lowpass','highpass','ranfilter2','eer_val','eer_eval']].groupby(['model','DataSize'])['eer_eval'].agg(**{
    'min_eer_eval':'min',
    'mean_eer_eval':'mean',
    'std_eer_eval' :'std'
}).reset_index()
df2
```

```python
df2_1 = df2.copy()
df2_1['model'] = df2_1['model'].apply(ModelRename)
df2_1['DataSize'] = df2_1['DataSize'].apply(Percentage)
df2_1['mean_eer_eval'] = df2_1['mean_eer_eval'].apply(Percentage)
```

```python
df2_1.to_csv("DataSize_HP.csv", index = False)
```

### HP + LP

```python
df1[df1.highpass != False][df1.lowpass != False][df1.ranfilter2 == False][['model','DataSize','lowpass','highpass','ranfilter2','eer_val','eer_eval']].sort_values(['model','DataSize','eer_eval'])
```

```python
df3 = df1[df1.highpass != False][df1.lowpass != False][df1.ranfilter2 == False][['model','DataSize','beta_param','lowpass','highpass','ranfilter2','eer_val','eer_eval']].groupby(['model','DataSize'])['eer_eval'].agg(**{
    'min_eer_eval':'min',
    'mean_eer_eval':'mean',
    'std_eer_eval' :'std'
}).reset_index()
df3
```

```python
df3_1 = df3.copy()
df3_1['model'] = df3_1['model'].apply(ModelRename)
df3_1['DataSize'] = df3_1['DataSize'].apply(Percentage)
df3_1['mean_eer_eval'] = df3_1['mean_eer_eval'].apply(Percentage)
df3_1
```
